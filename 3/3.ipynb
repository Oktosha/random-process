# Билет 3

**Ветвящиеся процессы Гальтона – Ватсона. Теорема о вероятности вырождения ветвящегося процесса.**

Ветвящийся процесс Гальтона-Ватсона и его вырождение определены в предыдущем билете (билете 2).

**Теорема (о вероятности вырождения)**. *Пусть $(X_n, n \in \mathbb{Z_+})$ — ветвящийся процесс Гальтона-Ватсона с законом размножения частиц $\xi$, причём $P(\xi = 1) \ne 1$. Тогда*:

1. *Если $E\xi \le 1$, то вероятность вырождения процесса равна 1. При этом 1 — единственное решение уравнения $z = \varphi(z)$*
2. *Если $E\xi > 1$, то вероятность вырождения процесса принадлежит интервалу $[0;1)$ и совпадает с единственным решением $z_0$ уравнения $z = \varphi(z)$ на этом интервале*

Как нам уже известно из предыдущего билета, вероятность вырождения процесса Гальтона-Ватсона с законом размножения частиц $\xi$ является решением уравнения $\varphi_\xi(z) = z$.

**Доказываем случай 1.** $E\xi \le 1$. По свойству производящих функций (см. билет 2) представим $\varphi_\xi$ в виде $\varphi_\xi(z) = \sum_{k = 0}^{+\infty}P(\xi = k)z^k$. Тогда её производная $\varphi_\xi'$ представится в виде $\varphi_\xi'(z)=\sum_{k = 1}^{+\infty}k P(\xi = k) z^{k - 1}$.

**Случай 1.a.** Если $\forall k > 1 \: P(\xi = k) = 0$ то это тривиальный случай: $\xi$ принимает значения только $0$ и $1$. В этом случае $P(X_n = 0) = 1 - (P(\xi = 1))^n$, Поскольку $P(\xi = 1) \ne 1$, то вероятность вырождения процесса равна $1$.

Докажем, что $1$ — единственное решение уравнения $\varphi_\xi(z) = z$. Действительно, если $\forall k > 1 \: P(\xi = k) = 0$, то $\varphi_\xi'(z) = a = \varphi_\xi'(1) = E\xi = P(\xi = 1) < 1$ и $\varphi_\xi(z) = az + b$. Таким образом, уравнение $\varphi_\xi(z) = z$ вырождается в линейное $(1-a)z = b$. Так как $a \ne 1$, то это уравнение имеет ровно одно решение. По свойству производящей функции $\varphi_\xi(1) = 1$. Таким образом, $1$ — единственное решение уравнения $\varphi_\xi(z) = z$.

**Случай 1.б.** Пусть $\exists k > 1$, такое что $P(\xi = k) \ne 0$. Тогда производная производящей функции строго возрастает.
